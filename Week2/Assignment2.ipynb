## Filtering the Dataset

```python
# Set a new variable 
desired_columns = ['PARK_NAME','GIS_ACRES','AGNCY_LEV','AGNCY_NAME','ACCESS_TYP','BSKTB','geometry']
```

```python
Parks_trimmed = Parks[desired_columns].copy()
Parks_trimmed
```

```python
# I want to see how many parks have have different types of access and know what the access types are.
Parks['ACCESS_TYP'].value_counts()
```

```python
# I want to create a new list of just the parks with Restricted Access
RParks = Parks_trimmed[Parks_trimmed.ACCESS_TYP == 'Restricted Access']
RParks
```

```python
# Lets see where the Restricted Access Parks are. 
RParks.plot()
```

```python
Parks.plot(
            figsize=(20,12),   #size of the plot
            column = 'ACCESS_TYP',   # Sort by Access Type
            legend = True,     # add a legend           
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1,0.3)
            }                  # Put the legend in the white space of the graph
) 
# Tried to use the function we used in class just to see how it would visualize the data. 
```
